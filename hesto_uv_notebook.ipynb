## Export report sub-tables
Each sub-table (publications, patents, student metrics, project status) is
written to its own CSV in **long format** — one row per item, with `project_id`
as the repeating link key back to `report_files.csv`. Only **active** reports
are included. Headers are normalized to clean names by column position.

```python
# Sub-table / master-table export logic lives in hestodb.export
from hestodb.export import (
    build_detail_csv,
    build_master_table,
    group_for,
    PUBLICATION_COLS,
    PATENT_COLS,
    STUDENT_COLS,
    STATUS_COLS,
)
```

## Master table (one row per active report)
Folds all six slide parsers into a single table (one row per active report):
- **summary** + **accommodation** columns come straight from `files`
- **project_status** is pivoted to fixed `{category}_prior / _current / _rationale` columns
- **publications / patents / student_metrics** contribute a count (`n_*`) plus a `" ;; "`-joined detail column (full detail lives in the per-table CSVs above)

```python
master = build_master_table(reports, files, "master_table.csv")
print(
    f"master_table.csv: {len(master)} rows, {master.shape[1]} columns "
    "(grouped 2-row header)"
)
print("read back with:  pd.read_csv('master_table.csv', header=[0, 1])")

# preview using the grouped header
master_preview = master.copy()
master_preview.columns = pd.MultiIndex.from_tuples(
    [(group_for(c), c) for c in master.columns]
)
master_preview.head()
```

## Smith test-file verification table
Collects **everything** parsed from the Smith test file (`test_report`) into one
tidy long-format dataframe so each parsed value can be checked against the source
slides. Columns: `section | row | field | value`. Scalar header fields and the
summary dict live on `row 0`; each parsed table contributes one row per
(record, column) cell.

```python
# Build one tidy long-format table of everything parsed from the Smith test file.
verify_rows = []


def _add_scalar(section, field, value):
    verify_rows.append({"section": section, "row": 0, "field": field, "value": value})


def _add_frame(section, df):
    if df is None or len(df) == 0:
        return
    df = df.reset_index(drop=True)
    for i, record in df.iterrows():
        for col in df.columns:
            verify_rows.append(
                {
                    "section": section,
                    "row": i,
                    "field": str(col),
                    "value": record[col],
                }
            )


# scalar header fields
_add_scalar("metadata", "project_id", test_report.project_id)
_add_scalar("metadata", "principal_investigator", test_report.principal_investigator)
_add_scalar("metadata", "affiliation", test_report.affiliation)
_add_scalar("metadata", "research_regime", test_report.research_regime)

# summary dict (one field per key)
for key, value in (test_report.summary or {}).items():
    _add_scalar("summary", key, value)

# parsed tables (one row per cell)
_add_frame("accommodation", test_report.accomodation_table)
_add_frame("publications", test_report.publication_table)
_add_frame("patents", test_report.patents_table)
_add_frame("student_metrics", test_report.student_metrics_table)
_add_frame("project_status", test_report.project_status)

smith_verify = pd.DataFrame(verify_rows, columns=["section", "row", "field", "value"])
print(
    f"smith_verify: {len(smith_verify)} parsed values across "
    f"{smith_verify['section'].nunique()} sections "
    f"({', '.join(smith_verify['section'].unique())})"
)

# show every row so values can be eyeballed against the slides
with pd.option_context("display.max_rows", None, "display.max_colwidth", None):
    display(smith_verify)
```
