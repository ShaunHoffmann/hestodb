## Export report sub-tables
Each sub-table (publications, patents, student metrics, project status) is
written to its own CSV in **long format** — one row per item, with `project_id`
as the repeating link key back to `report_files.csv`. Only **active** reports
are included. Headers are normalized to clean names by column position.

```python
# Sub-table / master-table export logic lives in hestodb.export
from hestodb.export import (
    build_detail_csv,
    build_master_table,
    group_for,
    PUBLICATION_COLS,
    PATENT_COLS,
    STUDENT_COLS,
    STATUS_COLS,
)
```

```python
publications = build_detail_csv(
    reports, files, "publication_table", PUBLICATION_COLS, "publications.csv"
)
print(
    f"publications.csv: {len(publications)} rows "
    f"({publications['project_id'].nunique()} projects)"
)
publications.head()
```

```python
patents = build_detail_csv(reports, files, "patents_table", PATENT_COLS, "patents.csv")
print(f"patents.csv: {len(patents)} rows ({patents['project_id'].nunique()} projects)")
patents.head()
```

```python
student_metrics = build_detail_csv(
    reports, files, "student_metrics_table", STUDENT_COLS, "student_metrics.csv"
)
print(
    f"student_metrics.csv: {len(student_metrics)} rows "
    f"({student_metrics['project_id'].nunique()} projects)"
)
student_metrics.head()
```

```python
project_status = build_detail_csv(
    reports, files, "project_status", STATUS_COLS, "project_status.csv"
)
print(
    f"project_status.csv: {len(project_status)} rows "
    f"({project_status['project_id'].nunique()} projects)"
)
project_status.head()
```

## Master table (one row per active report)
Folds all six slide parsers into a single table (one row per active report):
- **summary** + **accommodation** columns come straight from `files`
- **project_status** is pivoted to fixed `{category}_prior / _current / _rationale` columns
- **publications / patents / student_metrics** contribute a count (`n_*`) plus a `" ;; "`-joined detail column (full detail lives in the per-table CSVs above)

```python
master = build_master_table(reports, files, "master_table.csv")
print(
    f"master_table.csv: {len(master)} rows, {master.shape[1]} columns "
    "(grouped 2-row header)"
)
print("read back with:  pd.read_csv('master_table.csv', header=[0, 1])")

# preview using the grouped header
master_preview = master.copy()
master_preview.columns = pd.MultiIndex.from_tuples(
    [(group_for(c), c) for c in master.columns]
)
master_preview.head()
```

```python
frames = []


def add_table(df, label):
    if df is not None:
        df = df.copy()
        df["table"] = label
        frames.append(df)


add_table(test_report.student_metrics_table, "student_metrics")
add_table(test_report.accomodation_table, "accomodation")
add_table(test_report.publication_table, "publications")
add_table(test_report.patents_table, "patents")
add_table(test_report.project_status, "project_status")

combined = pd.concat(frames, ignore_index=True)
combined
```
